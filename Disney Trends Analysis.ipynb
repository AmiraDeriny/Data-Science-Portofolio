## 3 - Data transformation

Hints:

  - One hot encode the categorical columns, aka. 'genre' column

```python
# get dummies one One hot encode the categorical columns, aka. 'genre' column)
df = df.join(pd.get_dummies(df['genre']))
df
```

## 4- Model

Hints:

  - use sm.OLS
  - use X's as the one hot encoded genres, Y as the inflation_adjusted_gross

```python
df['intercept'] = 1
df.head()
```

```python
lm = sm.OLS(df['inflation_adjusted_gross'], df[['intercept', 'Action', 'Adventure', 'Black Comedy', 'Comedy', 'Concert/Performance', 'Documentary', 'Drama', 'Horror', 'Musical', 'Romantic Comedy', 'Thriller/Suspense']])
results = lm.fit()
results.summary()
```
